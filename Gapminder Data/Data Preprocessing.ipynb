```python
# tidy data format

pop_tidy = pd.melt(pop, ['country'], var_name='year', value_name='population')
pop_tidy = pop_tidy.sort_values(by=['country', 'year'])
pop_tidy
```

```python
gni_tidy = pd.melt(gni_fill, ['country'], var_name='year', value_name='GNI')
gni_tidy = gni_tidy.sort_values(by=['country', 'year'])
gni_tidy
```

```python
exp_tidy = pd.melt(exp, ['country'], var_name='year', value_name='life expectancy')
exp_tidy = exp_tidy.sort_values(by=['country', 'year'])
exp_tidy
```

```python
data = pop_tidy.merge(exp_tidy).merge(gni_tidy)
data
```
